```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

```python
train=np.load('datasets/train/train_data.npy')
test=np.load('datasets/test/test_data.npy')
```

```python
X_train=np.array(train[:,:-1])
y_train=np.array(train[:,-1:])

X_test=np.array(test[:,:-1])
y_test=np.array(test[:,-1:])
```

```python
model = Sequential()

model.add(Dense(8, input_dim=8, activation='tanh'))
model.add(Dense(8, activation='tanh'))
model.add(Dense(4, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))

sgd=SGD(lr=0.01)
model.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['acc'])
```

```python
history=model.fit(X_train, y_train, epochs=100, verbose=2, validation_split=0.2)
```

```python
score=model.evaluate(X_test, y_test)
print(model.metrics_names[0], '%.6f' % (score[0]))
print(model.metrics_names[1], score[1]*100, '%')
```
